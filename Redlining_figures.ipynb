```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors
from google.colab import files
import matplotlib as mpl
import matplotlib.ticker as mtick
from matplotlib.ticker import FuncFormatter
import warnings
warnings.filterwarnings('ignore')
```

```python
# Violin plot between income and tree cover: Baltimore 2020

inc = pd.read_csv("/content/drive/My Drive/balt_holc_inc_2020.csv")
inc = inc[inc['grade'].notna()]

inc = inc[inc['total_treey'].notna()]
inc = inc.sort_values(['grade'])
cols = ['#8FB780','#94C2C6','#F0E771','#D06B69']

fig, axes = plt.subplots(figsize=(5, 4))
sns.violinplot(x ='grade', y ='total_treey', data = inc, palette=cols)

df_means = inc.groupby("grade")["total_treey"].agg("mean").reset_index()
xlim = axes.get_xlim()
ylim = axes.get_ylim()

plt.title('', fontsize=12)
axes.spines[['right', 'top']].set_visible(False)

plt.xlabel('HOLC Grade',size=12)
plt.ylabel('% Total Tree Cover',size=12)
for tick in axes.xaxis.get_major_ticks():
    tick.label.set_fontsize(12)

for tick in axes.yaxis.get_major_ticks():
    tick.label.set_fontsize(12)
axes.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
axes.set(ylim=(-.1, 1))
plt.savefig("balt_redline_tree_v.jpg",dpi=300,bbox_inches='tight')
files.download("balt_redline_tree_v.jpg")
```
